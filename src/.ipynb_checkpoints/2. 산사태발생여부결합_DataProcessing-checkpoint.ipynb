```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, LineString
from fiona.crs import from_string

epsg4326 = from_string("+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs")
epsg5179 = from_string("+proj=tmerc +lat_0=38 +lon_0=127.5 +k=0.9996 +x_0=1000000 +y_0=2000000 +ellps=GRS80 +units=m +no_defs")
```

Spatial join the two shape files

```python
admin_boundary = gpd.GeoDataFrame.from_file('/Users/ohikendoit/PycharmProjects/weather_bigcon_2021/data/Admin_Boundary/umd.shp', encoding='utf-8')
#admin_boundary = gpd.GeoDataFrame.from_file('/Users/ohikendoit/PycharmProjects/weather_bigcon_2021/data/landslide_events.csv', encoding='euc-kr')
admin_boundary.crs = epsg4326
#soil_information = gpd.GeoDataFrame.from_file('/Users/ohikendoit/PycharmProjects/weather_bigcon_2021/data/soil_information.csv', encoding='euc-kr')
#soil_information.crs = epsg4326
soil_47_kb_1 = gpd.GeoDataFrame.from_file('/Users/ohikendoit/PycharmProjects/weather_bigcon_2021/data/Soil_47_KB/TB_FGDI_FS_IJ5000_PG_47_1.shp', encoding='utf-8')
soil_47_kb_1.crs = epsg4326
soil_47_kb_2 = gpd.GeoDataFrame.from_file('/Users/ohikendoit/PycharmProjects/weather_bigcon_2021/data/Soil_47_KB/TB_FGDI_FS_IJ5000_PG_47_2.shp', encoding='utf-8')
soil_47_kb_2.crs = epsg4326
soil_48_kn = gpd.GeoDataFrame.from_file('/Users/ohikendoit/PycharmProjects/weather_bigcon_2021/data/Soil_48_KN/TB_FGDI_FS_IJ5000_PG_48.shp', encoding='utf-8')
soil_48_kn.crs = epsg4326
```

```python
admin_boundary["geometry"] = admin_boundary["geometry"].buffer(5)
result_soil_47_kb_1 = gpd.sjoin(soil_47_kb_1, admin_boundary, how='left', op='within')
result_soil_47_kb_1.head()
```

```python
admin_boundary["geometry"] = admin_boundary["geometry"].buffer(5)
result_soil_47_kb_2 = gpd.sjoin(soil_47_kb_2, admin_boundary, how='left', op='within')
result_soil_47_kb_2.head()
```

```python
admin_boundary["geometry"] = admin_boundary["geometry"].buffer(5)
result_soil_48_kn = gpd.sjoin(soil_48_kn, admin_boundary, how='left', op='within')
result_soil_48_kn.head()
```

```python
soil_information = pd.concat([result_soil_47_kb_1, result_soil_47_kb_2, result_soil_48_kn])
#soil_information.to_csv("soil_information_derived.csv", encoding='utf-8')
```
