```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, LineString
from fiona.crs import from_string
```

```python
soil_information = pd.read_csv("soil_information_labelled.csv", encoding='utf-8')
forest_information = pd.read_csv("forest_information_labelled.csv", encoding='utf-8')
```

```python
soil_information = soil_information.drop(soil_information.columns[0], axis=1)
forest_information = forest_information.drop(forest_information.columns[0], axis=1)
```

```python
soil_information = soil_information.groupby('ADM_DR_NM', as_index=False).mean()
soil_information
```

```python
forest_information = forest_information.groupby('ADM_DR_NM', as_index=False).mean()
forest_information
```

앙상블 기반의 랜덤포레스트를 통해 산사태 발생여부에 미치는 영향을 결정
- 결과를 본다음 그룹바이를 할지 하지말지 판단해야함

```python
from sklearn.model_selection import train_test_split

x = soil_information[['PRRCK_LARG','PRRCK_MDDL','LOCTN_ALTT','LOCTN_GRDN','EIGHT_AGL','CLZN_CD','TPGRP_TPCD','PRDN_FOM_C','SLANT_TYP','SLDPT_TPCD','SCSTX_CD','SLTP_CD','ADM_DR_CD']]
y = soil_information['landslide']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

forest = RandomForestClassifier(n_estimators=100,
                               max_features='auto',
                               bootstrap=True,
                               )
forest.fit(x_train, y_train)

y_pred = forest.predict(x_test)
print(y_pred)
print(list(y_test))

print('accuracy:', metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from dtreeviz.trees import dtreeviz

visualization = dtreeviz(forest.estimators_[99], x, y,
                        target_name='landslide occurence',
                        feature_names=list(x),
                        class_names=list(x),
                        title="Random forest of Landslide Data")
visualization.save("decision_tree_soil.svg")
```
